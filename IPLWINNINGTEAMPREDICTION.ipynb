```python
match_df['team1']= match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']= match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x']- delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['balls_left']  = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
#wicket left
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else "1")
```

```python
delivery_df.tail()
```

```python
delivery_df['player_dismissed'].dtypes
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wicket = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
wicket
```

```python
delivery_df['wicket'] = 10 - wicket
```

```python
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
#current run rate
delivery_df['crr']= (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
#required run rate - rrr
delivery_df['rrr']= (delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df.tail()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
#traintestsplit
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2)
```

```python
#standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
#ml model - logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc = accuracy_score(y_pred,y_test)
rec = recall_score(y_pred,y_test)
pre = precision_score(y_pred,y_test)
f1 = f1_score(y_pred,y_test)
print(acc,rec,pre,f1)
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error,r2_score
msq = mean_squared_error(y_pred,y_test)
r2 = r2_score(y_pred,y_test)
print("Mean sqaure error is:",msq)
print("r2 score is:",r2)
```
